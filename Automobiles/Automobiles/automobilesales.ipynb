```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv("Auto_Sales_data.csv")
df1
```

```python
df1.describe()
```

```python
df1.info()
```

```python
df1['ORDERDATE'] = pd.to_datetime(df1['ORDERDATE'])
df1['ORDERDATE']
```

```python
df1["PRODUCTCODE"].unique()
```

```python
df1["PRODUCTCODE"] = df1["PRODUCTCODE"].apply(lambda x: float(''.join(filter(str.isdigit, x))))
df1["PRODUCTCODE"]
```

```python
df1["PHONE"].unique()
```

```python
df1["PHONE"] = df1["PHONE"].apply(lambda x: int(''.join(filter(str.isdigit, x))))
df1["PHONE"]
```

```python
df1['POSTALCODE'] = pd.to_numeric(df1['POSTALCODE'], errors='coerce').astype(float)

df1['POSTALCODE']
```
