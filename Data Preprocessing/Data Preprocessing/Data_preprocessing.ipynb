### Replace all the values containing special characters and numbers with NaN

```python
# ord('Z')
# chr(122)


valid_chars=[" "]
valid_chars.extend([chr(ord('A')+i) for i in range(0,25)])
valid_chars.extend([chr(ord('A')+i) for i in range(0,25)]) 
print(valid_chars)
```

```python
def replace_special_chars(s):
    if not isinstance(s,str):
        return s
    for c in s:
        if c not in valid_chars:
            return np.nan
        return s
df["Shape Reported"]=df["Shape Reported"].apply(replace_special_chars)

## Replace NaN with 'Point'
df["Shape Reported"].fillna('Point',inplace=True)
df
```
