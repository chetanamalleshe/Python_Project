### 10. Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos.

### 11. Find the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences.

```python
# The title of the video which has the maximum number of likes
maxLikes=df["likes"].max()
maxLikes_title=df[df["likes"]== maxLikes]['title'].values[0]
print(f"The title of the video which has the maximum likes is: {maxLikes_title}")
```

```python
# The title of the video which has the minimum number of likes
minLikes=df["likes"].min()
minLikes_title=df[df["likes"]== minLikes]['title'].values[0]
print(f"The title of the video which has the minimum likes is: {minLikes_title}")
```

### Inferences:

- "BTS 'Dynamite' Official MV" is the most liked video in the dataset.This suggests that the video is exceptionally popular.
- "Kim Kardashian's Must-See Moments on 'Saturday Night Live' | E! News" has the minimum number of likes, indicating that it may not have received as much attention compared to other videos in the dataset.

### 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.


```python
correlation = df["view_count"].corr(df["dislikes"])
correlation
```

```python
#creating Scatter plot
import matplotlib.pyplot as plot
```

```python
X = df["view_count"]
Y = df["dislikes"]
plot.scatter(X,Y,c= 'green')
plot.xlabel("Views")
plot.ylabel("Dislikes")
plot.title("Correlation between Views and Dislikes")
plot.show()
```

### Inferences

- The correlation between views and dislikes is 0.684.
Suggests a positive correlation between the number of views and the number of dislikes on a video.
- This means that as the view count increases, the number of dislikes tends to increase as well.
and other factors may influence the number of dislikes on a video, such as the video's content,quality etc.
